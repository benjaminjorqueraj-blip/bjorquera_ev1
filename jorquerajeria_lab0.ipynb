## 8. Tu tarea: completar la implementación

Se debe implementar los siguientes métodos:

- `obtener(posicion)`
- `eliminar(posicion)`
- `mostrar_todos()`
- `calcular_saldo()`

### Recomendaciones
- validar si la posición existe
- decidir cuándo conviene usar `return`
- probar cada método apenas se complete

```python
class ListaMovimientos:
    def __init__(self):
        self.movimientos = []

    def esta_vacia(self):
        return len(self.movimientos) == 0

    def agregar(self, movimiento):
        self.movimientos.append(movimiento)

    def tamano(self):
        return len(self.movimientos)

    def obtener(self, posicion):
        # Validamos que la posición exista antes de acceder
        if posicion < 0 or posicion >= len(self.movimientos):
            print("Error: posición", posicion, "no existe. El rango válido es 0 a", len(self.movimientos) - 1)
            return None
        return self.movimientos[posicion]

    def eliminar(self, posicion):
        # Validamos que la posición exista antes de eliminar
        if posicion < 0 or posicion >= len(self.movimientos):
            print("Error: posición", posicion, "no existe. No se eliminó ningún movimiento.")
            return None
        eliminado = self.movimientos[posicion]
        self.movimientos.pop(posicion)
        print("Movimiento eliminado:", eliminado["descripcion"])
        return eliminado

    def mostrar_todos(self):
        if self.esta_vacia():
            print("La lista no tiene movimientos.")
            return
        print("--- Movimientos de la cuenta ---")
        for i in range(len(self.movimientos)):
            m = self.movimientos[i]
            print(i, "|", m["descripcion"], "| $" + str(m["monto"]), "|", m["tipo"])
        print("--------------------------------")

    def calcular_saldo(self):
        saldo = 0
        for m in self.movimientos:
            if m["tipo"] == "abono":
                saldo = saldo + m["monto"]
            elif m["tipo"] == "cargo":
                saldo = saldo - m["monto"]
        return saldo
```

## 10. Desafío opcional

Si se avanza rápido, implementar una o más de estas mejoras:

1. Buscar movimientos por una palabra en la descripción
2. Obtener solo los cargos sobre un cierto monto
3. Crear una clase `Movimiento`
4. Validar que el tipo solo pueda ser `"abono"` o `"cargo"`
5. Crear un resumen con:
   - total de abonos
   - total de cargos
   - saldo final
   - cantidad de movimientos

```python
# Desafío opcional

class ListaMovimientosExtendida(ListaMovimientos):

    def buscar_por_descripcion(self, palabra):
        # busca movimientos que tengan esa palabra en la descripcion
        resultados = []
        for m in self.movimientos:
            if palabra.lower() in m["descripcion"].lower():
                resultados.append(m)
        return resultados

    def cargos_mayores_a(self, monto_minimo):
        # retorna solo los cargos que sean mayores al monto indicado
        resultado = []
        for m in self.movimientos:
            if m["tipo"] == "cargo" and m["monto"] > monto_minimo:
                resultado.append(m)
        return resultado

    def resumen(self):
        total_abonos = 0
        total_cargos = 0
        for m in self.movimientos:
            if m["tipo"] == "abono":
                total_abonos = total_abonos + m["monto"]
            elif m["tipo"] == "cargo":
                total_cargos = total_cargos + m["monto"]
        print("===== RESUMEN =====")
        print("Movimientos:", self.tamano())
        print("Total abonos: $" + str(total_abonos))
        print("Total cargos: $" + str(total_cargos))
        print("Saldo: $" + str(total_abonos - total_cargos))


cuenta2 = ListaMovimientosExtendida()
cuenta2.agregar({"descripcion": "Deposito sueldo", "monto": 850000, "tipo": "abono"})
cuenta2.agregar({"descripcion": "Pago arriendo", "monto": 320000, "tipo": "cargo"})
cuenta2.agregar({"descripcion": "Compra farmacia", "monto": 18990, "tipo": "cargo"})
cuenta2.agregar({"descripcion": "Transferencia recibida", "monto": 50000, "tipo": "abono"})
cuenta2.agregar({"descripcion": "Cobro mantencion", "monto": 6500, "tipo": "cargo"})

print("Busqueda 'pago':", cuenta2.buscar_por_descripcion("pago"))
print()
print("Cargos mayores a 10000:", cuenta2.cargos_mayores_a(10000))
print()
cuenta2.resumen()
```
